```python
#Построить 3д-график в виде сферы с радиусом 0.5 см

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='summer')
ax.legend()
```
